```python
import java.sql.*;
import java.util.*;

public class OnlineQuiz {
    private static final String DB_URL = "jdbc:mysql://localhost:3306/quiz_db";
    private static final String DB_USER = "your_username";
    private static final String DB_PASSWORD = "your_password";

    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);
        Connection connection = null;
        Statement statement = null;
        ResultSet resultSet = null;

        try {
            // Establish connection to the database
            connection = DriverManager.getConnection(DB_URL, DB_USER, DB_PASSWORD);
            statement = connection.createStatement();

            // Main loop for user interaction
            while (true) {
                System.out.println("Choose an option:");
                System.out.println("1. Login");
                System.out.println("2. Exit");
                int choice = scanner.nextInt();

                switch (choice) {
                    case 1:
                        // Login form
                        System.out.print("Enter username: ");
                        String username = scanner.next();
                        System.out.print("Enter password: ");
                        String password = scanner.next();

                        // Verify login credentials
                        String query = "SELECT * FROM users WHERE username='" + username + "' AND password='" + password + "'";
                        resultSet = statement.executeQuery(query);

                        if (resultSet.next()) {
                            // Successful login
                            System.out.println("Login successful!");

                            // Display main menu
                            while (true) {
                                System.out.println("Choose an option:");
                                System.out.println("1. Update Profile");
                                System.out.println("2. Update Password");
                                System.out.println("3. Start Quiz");
                                System.out.println("4. Logout");
                                choice = scanner.nextInt();

                                switch (choice) {
                                    case 1:
                                        // Update Profile form
                                        System.out.print("Enter new name: ");
                                        String newName = scanner.next();
                                        System.out.print("Enter new email: ");
                                        String newEmail = scanner.next();

                                        // Update profile details in the database
                                        query = "UPDATE users SET name='" + newName + "', email='" + newEmail + "' WHERE username='" + username + "'";
                                        statement.executeUpdate(query);
                                        System.out.println("Profile updated successfully!");
                                        break;
                                    case 2:
                                        // Update Password form
                                        System.out.print("Enter current password: ");
                                        String currentPassword = scanner.next();
                                        System.out.print("Enter new password: ");
                                        String newPassword = scanner.next();

                                        // Verify current password
                                        query = "SELECT * FROM users WHERE username='" + username + "' AND password='" + currentPassword + "'";
                                        resultSet = statement.executeQuery(query);

                                        if (resultSet.next()) {
                                            // Update password in the database
                                            query = "UPDATE users SET password='" + newPassword + "' WHERE username='" + username + "'";
                                            statement.executeUpdate(query);
                                            System.out.println("Password updated successfully!");
                                        } else {
                                            System.out.println("Invalid current password.");
                                        }
                                        break;
                                    case 3:
                                        // Start Quiz
                                        // Implement quiz logic here (e.g., fetch questions from database, display questions, handle user answers, calculate score)
                                        break;
                                    case 4:
                                        // Logout
                                        System.out.println("Logging out...");
                                        break;
                                    default:
                                        System.out.println("Invalid choice.");
                                }

                                if (choice == 4) {
                                    break;
                                }
                            }
                        } else {
                            System.out.println("Invalid login credentials.");
                        }
                        break;
                    case 2:
                        // Exit the program
                        System.out.println("Exiting...");
                        System.exit(0);
                    default:
                        System.out.println("Invalid choice.");
                }
            }
        } catch (SQLException e) {
            e.printStackTrace();
        } finally {
            // Close resources
            try {
                if (resultSet != null) {
                    resultSet.close();
                }
                if (statement != null) {
                    statement.close();
                }
                if (connection != null) {
                    connection.close();
                }
            } catch (SQLException e) {
                e.printStackTrace();
            }
        }
    }
}
```
